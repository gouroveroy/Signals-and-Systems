```python
import numpy as np
import matplotlib.pyplot as plt

# Define the signal x[n] for example
x = np.array([0, 1, 2, 3, 4, 3, 2, 1, 0, -1, -2])  # Example signal

# Time index n (must cover both negative and positive indices for this example)
n = np.arange(-5, 6)  # n from -5 to 5


# Create the impulse function [n-k]
def delta(n, k):
    return np.where(n == k, 1, 0)


# Create a figure with multiple subplots (4 rows, 3 columns)
fig, axs = plt.subplots(4, 3, figsize=(12, 10))

# Title for the entire figure
fig.suptitle("Impulses multiplied by coefficients", fontsize=16)

# Plot the individual impulses [n-k] * x[k]
for i, k in enumerate(n):
    row = i // 4
    col = i % 3
    axs[row, col].stem(n, delta(n, k) * x[k], basefmt="r-")
    axs[row, col].set_ylim(-1, 4)
    axs[row, col].set_title(f"[n - ({k})]x[{k}]")
    axs[row, col].set_xlabel("n (Time Index)")
    axs[row, col].set_ylabel("x[n]")
    axs[row, col].grid(True)

# Plot the sum of all impulse responses in the last subplot
summed_signal = np.zeros_like(n)
for k in n:
    summed_signal += delta(n, k) * x[k]

axs[3, 2].stem(n, summed_signal, basefmt="r-")
axs[3, 2].set_ylim(-1, 4)
axs[3, 2].set_title("Sum")
axs[3, 2].set_xlabel("n (Time Index)")
axs[3, 2].set_ylabel("x[n]")
axs[3, 2].grid(True)

# Adjust layout to prevent overlapping of plots
plt.tight_layout()
plt.subplots_adjust(top=0.9)

# Display the plot
plt.show()
```
